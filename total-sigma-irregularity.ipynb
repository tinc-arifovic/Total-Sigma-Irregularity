```python
def sigma1 (G):
    n = order(G)
    return sum((abs(u-v)) ** (1/n) for u, v in combinations(G.degree(),2))
```

```python
def sigma2 (G,c):
    n = order(G)
    return sum((abs(u-v)) ** c/100 for u, v in combinations(G.degree(),2))
```

```python
def sigma3 (G): # dodati je treba izbrane funkcije
    n = order(G)
    return sum((abs(u-v)) ** (1/n) for u, v in combinations(G.degree(),2))
```

```python
def indeksi_maksimalnih_elementov(seznam):
    maks_vrednost = max(seznam)
    indeksi = [i for i, vrednost in enumerate(seznam) if vrednost == maks_vrednost]

    return indeksi
```
